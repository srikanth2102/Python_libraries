```python
# importing opnecv
import cv2 as cv
```

```python
# reading an image from disk
im = cv.imread("F:\Desktop\WhatsApp Image 2021-08-29 at 10.37.48.jpeg")
```

```python
# importing matplotlib
from matplotlib import pyplot as plt
```

```python
# plotting the imported image
plt.imshow(im)
plt.show()
```

```python
# converting the BGR image to HSV
HSV_im = cv.cvtColor(im, cv.COLOR_BGR2HSV)
```

```python
# plotting the HSV image
plt.imshow(HSV_im)
plt.show()
```

```python
# splitting the Hue, Saturation and Value
Hue, Saturation, Value = cv.split(HSV_im)
```

```python
# plotting the Hue
plt.imshow(Hue)
plt.show()
```

```python
# plotting the saturation channel
plt.imshow(Saturation)
plt.show()
```

```python
# plotting the value channel
plt.imshow(Value)
plt.show()
```

```python
# finding the brightness value
Value.mean()
```

```python
# importing a new bright image
new_im = cv.imread("F:\Desktop\WhatsApp Image 2021-08-29 at 10.37.49.jpeg")
```

```python
# plotting the new bright image
plt.imshow(new_im)
plt.show()
```

```python
# creating a function to find the brightness
def brightness(im):
    HSV_im = cv.cvtColor(im, cv.COLOR_BGR2HSV)
    
    Hue, Saturation, Value = cv.split(HSV_im)
    
    return(Value.mean())
```

```python
# brightness of the first image
print("Brightness of first image",brightness(im))
```

```python
# brightness of the second image
print("Brightness of second image", brightness(new_im))
```

```python
# telling the webcam to start recording. The parameter '0' indicated that the video will be captured using the webcam

vid = cv.VideoCapture(0)
```

```python
# getting ten frames of image from webcam
# we can get continuosly get frame from webcame by changing the for loop to an infinite while loop
### like while(True):{....}

# to display the image along with its brightness
for i in range(10):
    ret, frame = vid.read()
    print(brightness(frame))
    plt.imshow(frame)
    plt.show()
```

```python
# displaying the captured image along with the brightness value
for i in range(10):
    ret, frame = vid.read()
    
    # we can use some conditon like this to check if the brightness is less and if it is less we can start the image detection using deep learning models.
    if (brightness(frame)<60):
        print(brightness(frame))
        plt.imshow(frame)
        plt.show()
```

```python
# stopping the recording of the webcam
vid.release()
```
